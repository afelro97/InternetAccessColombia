```python
import pandas as pd
```

```python
# Lista de archivos
files = [
    'INFORMACIÓN_TRIMESTRAL_DE_ACCESOS_FIJOS_A_INTERNET_POR_PROVEEDOR,_DEPARTAMENTO,_MUNICIPIO,_SEGMENTO,_TECNOLOGIA,_Y_VELOCIDAD_DE_CONEXIÓN_4.csv',
    'INFORMACIÓN_TRIMESTRAL_DE_ACCESOS_FIJOS_A_INTERNET_POR_PROVEEDOR,_DEPARTAMENTO,_MUNICIPIO,_SEGMENTO,_TECNOLOGIA,_Y_VELOCIDAD_DE_CONEXIÓN_5.csv',
    'INFORMACIÓN_TRIMESTRAL_DE_ACCESOS_FIJOS_A_INTERNET_POR_PROVEEDOR,_DEPARTAMENTO,_MUNICIPIO,_SEGMENTO,_TECNOLOGIA,_Y_VELOCIDAD_DE_CONEXIÓN_6.csv',
    'INFORMACIÓN_TRIMESTRAL_DE_ACCESOS_FIJOS_A_INTERNET_POR_PROVEEDOR,_DEPARTAMENTO,_MUNICIPIO,_SEGMENTO,_TECNOLOGIA,_Y_VELOCIDAD_DE_CONEXIÓN_7.csv'
]
```

```python
# Función para leer y mostrar las columnas de cada archivo
def verificar_columnas(files):
    for file in files:
        try:
            # Leer el archivo CSV
            df = pd.read_csv(file, nrows=5)  # Solo leer las primeras filas para ver las columnas
            print(f"Archivo: {file}")
            print(f"Columnas: {df.columns.tolist()}\n")
        except Exception as e:
            print(f"Error al leer {file}: {e}")
```

```python
# Ejecutar la función
verificar_columnas(files)
```

```python
# Lista para almacenar cada DataFrame
dataframes = []
```

```python
# Leer cada archivo CSV y agregarlo a la lista de DataFrames
for file in files:
    df = pd.read_csv(file)
    dataframes.append(df)
```

```python
# Combinar todos los DataFrames en uno solo
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
# Guardar el DataFrame combinado en un archivo CSV
combined_df.to_csv('INFORMACIÓN_TRIMESTRAL_DE_ACCESOS_FIJOS_A_INTERNET_POR_PROVEEDOR,_DEPARTAMENTO,_MUNICIPIO,_SEGMENTO,_TECNOLOGIA,_Y_VELOCIDAD_DE_CONEXIÓN_COMPLETO.csv', index=False)
```
